```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# hyperparameters

block_size = 3 # the number of words we use for our context
lr = -0.1 # learning rate
dimensionality = 2 # the number of dimensions in our model
n_neurons_hidden_layer = 100
dictionary_size = 27
batch_size = 32
```

```python
words = open('names.txt', 'r').read().splitlines() #load our dataset
words[:1]
```

```python
stoi = {s: i+1 for i, s in enumerate(sorted(list(set(''.join(words)))))}
stoi['.'] = 0
itos = {i: s for s, i in stoi.items()} #create our dictionary
```

```python
def build_dataset(words):
    block_size = 3
    X, Y = [], []

    for w in words:
        context = [0] * block_size
        for ch in w + '.':
            ix = stoi[ch]
            X.append(context)
            Y.append(ix)

            context = context[1:] + [ix]

    X = torch.tensor(X)
    Y = torch.tensor(Y)
    print(X.shape, Y.shape) # should be (vocab_size, block_size) and (vocab_size)
    return X, Y

import random
random.seed(42)
random.shuffle(words)
n1 = int(0.8*len(words))
n2 = int(0.9*len(words))

Xtr, Ytr = build_dataset(words[:n1])
Xval, Yval = build_dataset(words[n1:n2])
Xtest, Ytest = build_dataset(words[n2:])
```

```python
g = torch.Generator().manual_seed(2147483647)
C = torch.randn((27, 2), generator=g)

# Weights and biases for our first layer
W1 = torch.randn((block_size * dimensionality, n_neurons_hidden_layer), generator=g)
b1 = torch.randn(n_neurons_hidden_layer, generator=g)

# Weights and biases for our second layer
W2 = torch.randn((n_neurons_hidden_layer, dictionary_size), generator=g)
b2 = torch.randn(dictionary_size, generator=g)

parameters = [C, W1, b1, W2, b2] # all pointers. python is like scheme; all is pointer.

for p in parameters:
    p.requires_grad = True
```

```python
stepi = []
lossi = []
```

```python
def step(i):
    ix = torch.randint(0, Xtr.shape[0], (batch_size, ), generator = g)

    emb = C[Xtr[ix]] 
    h = torch.tanh(emb.view(-1, 6) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Ytr[ix])

    # zero grad
    for p in parameters:
        p.grad = None

    loss.backward()

    for p in parameters:
        p.data += lr * p.grad

    stepi.append(i)
    lossi.append(loss.item())
    
    return ix
```

```python
for i in range(10000, 20000):
    step(i)
plt.plot(stepi, lossi) 
```

```python
emb = C[Xval]
h = torch.tanh(emb.view(-1, 6) @ W1 + b1)
logits = h @ W2 + b2
loss = F.cross_entropy(logits, Yval)
loss
```

```python
def evaluate(inputs, outputs):
    emb = C[inputs] 
    h = torch.tanh(emb.view(-1, 6) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, outputs)
    return loss
```

```python
losstr = evaluate(Xtr, Ytr)
lossval = evaluate(Xval, Yval)
losstest = evaluate(Xtest, Ytest)

print(f"Training loss: {losstr}")
print(f"Vaidation loss: {lossval}")
print(f"Testing loss: {losstest}")
```
